# 이미지로 연습하는 numpy

`PIL` 모듈은 이미지를 제어할 수 있는 기능을 제공합니다. `open` 함수로 저장된 이미지 데이터를 읽어 올 수 있습니다. img에는 `PngImageFile` 객체가 저장됩니다. 

```python
from PIL import Image
img = Image.open("/content/drive/MyDrive/Colab Notebooks/2022-04-18 Python/3.numpy/image/0.png")
img = img.convert('L')
img
```

- L : 흑백
- RGB : Red, Green, Blue 삼원색을 이용한 색 표현

`transpose` 메서드로 `flip`할 수있음

`PngImageFile` 객체를 `ndarray`로 변환

```python
import numpy as np

arr = np.array(img)
print(type(arr))
print(arr.shape)
```

```python
arr
```

ndarray는 matplot 라이브러리로 시각화할 수 있습니다. 

```python
from matplotlib import pyplot as plt
plt.imshow(arr, cmap='gray')
```
