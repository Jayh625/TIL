# 반복문
반복문은 자료구조의 데이터 수만큼 코드를 반복실행 합니다.  
자료구조에 있는 "데이터를 하나씩 바인딩하고 들여쓰기된 코드를 실행" 과정을 반복하는 겁니다. 
  
## 반복문 - 기본형
`for 변수 in 자료구조 : ` 형태로 코드를 작성합니다. 코드 끝에 콜론 `:` 있음에 주의하세요. 분기문과 같이 콜론 다음에는 들여쓰기와 함께 영향을 받는 코드를 작성합니다. 

<img src="https://i.ibb.co/gR952tp/for-0.png" width="300" style="float:left" />

###### 반복문을 사용하지 않고 한 줄에 하나의 값 출력하기

````python
리스트 = [0,1,2,3]
print(리스트[0])
print(리스트[1])
print(리스트[2])
print(리스트[3])
````

###### 반복문으로 코드를 정리하기

````python
for 변수 in 리스트:
  print(변수)
````

###### 연습문제 - 1
다음 코드의 실행 결과를 예측하라.

```python
리스트 = ["가", "나", "다"]
for 변수 in 리스트 :
    print(변수)
    print("-" * 10)
for 변수 in 리스트 :
    print(변수)
```

````python
리스트 = ["가", "나", "다"]
for i in 리스트 :
  print(i)
  print("-" * 10)
for i in 리스트 : 
  print(i)  
````

```python
리스트 = ["가", "나", "다"]
for 변수 in 리스트 :
    print(변수)    
    for k in 리스트 :
        print(k)
```

````python
리스트 = ["가", "나", "다"]
for i in 리스트 :
  print(i)
  for k in 리스트 : 
    print(k)  
  print("")
# i=0,k=0/k=1/k=2 : 가가나다
# i=1,k=0/k=1/k=2 : 나가나다
# i=2,k=0/k=1/k=2 : 다가나다
````

###### 연습문제 - 2
리스트에서 앞의 두 개의 값을 다음과 같이 출력하라.  
```
가
나
```

````python
# 슬라이싱 이용해서 문제를 해결하라.
리스트 = ["가", "나", "다"]
````

````python
for i in 리스트[:2]:
  print(i)
````

````python
# 분기문을 이용해서 문제를 해결하라.
리스트 = ["가", "나", "다"]
for i in 리스트:
  if i !="다":
    print(i)
````

````python
리스트 = ["가", "나", "다"]
num=0
for i in 리스트:
  num += 1 
  if num < 3:
    print(i)
````

###### 연습문제 - 3
20 이상인 값을 출력하라

````python
리스트 = [20, 10, 30]
for 변수 in 리스트:
    if 변수 >= 20:
        print(변수)
````

````python
#20이상인 값을 리스트에 저장하라 
리스트 = [20, 10, 30]
r=[]
for 변수 in 리스트:
    if 변수 >= 20:
       r.append(변수)
r
````

###### 연습문제 - 5
반복문을 사용해서 변수에 저장된 종목을 화면에 출력하라.  
```
보유종목:Naver 
보유종목:Samsung
보유종목:SK Hynix
```

````python
interest_stocks = ["Naver", "Samsung", "SK Hynix"]
````

````python
for i in interest_stocks:
  print("보유종목:"+i)
````

###### 연습문제 - 6
반복문을 사용해서 변수에 저장된 종목을 화면에 출력하라.  
```
보유종목:Naver 
--------------
보유종목:Samsung
--------------
보유종목:SK Hynix
--------------
```

````python
interest_stocks = ["Naver", "Samsung", "SK Hynix"]
for i in interest_stocks:
  print("보유종목:"+i)
  print("-"*14)
````

###### 연습문제 - 7
반복문을 사용해서 변수에 저장된 종목을 화면에 출력하라.   
```
1999년 성적: 수 
2000년 성적: 양 
2001년 성적: 미 
2002년 성적: 수 
```

````python
my_grade = ["1999-수", "2000-양", "2001-미", "2002-수"]
for i in my_grade:
  print(i.split("-")[0]+"년 성적:", i.split("-")[1])
````
