### 리스트의 데이터 삭제

`remove` 메서드는 삭제할 값을 입력받아 리스트에서 제거합니다. 

````python
list.remove("D")
list
````

`pop` 메서드는 인덱스를 입력받아 하나의 값을 제거합니다. 

````python
list.pop(0)
````

`del` 키워드를 사용해서 제거할 수도 있습니다. 

````python
del list[0]
````

`clear` 메서드로 모든 데이터를 제거할 수 있습니다. 

````python
list.clear()
````
