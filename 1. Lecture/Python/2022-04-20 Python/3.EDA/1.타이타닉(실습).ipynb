```python
def func(x) :
  if x == 0:
    return "없음"
  elif x == 1:
    return "보통"
  else:
    return "많음"

df['보호자'] = df['SibSp'].map(  func )
df.head()
```

```python
how = {'Survived': np.mean}
df.groupby('보호자').agg(how)
```

```python
df.groupby(['보호자', 'Sex']).agg(how)
```

```python
def func(x):
  if x < 8:
    return "가성비짱"
  elif x < 13:
    return "쏘쏘"
  elif x < 28:
    return "그저그래"
  else:
    return "호구"

df['Fare'].map(func)
```
