#### 유사도(Proximity)

1. Euclidean distance

2. Cosine Similarity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([0, 0, 0])
```

```python
L = [1, 2, 3]
L2 = [0, 0, 0]

for i, i2 in zip(L, L2):
    print(i + i2)

L + L2 # concatenation
```

```python
# numpy broadcasting
v1 + v2 # (1, 2, 3) + (0, 0, 0) = (1, 2, 3) ## vector addition
v2 - v1 # vector subtraction
v1 * v2 # elementwise multiplication
v2 / v1 # elementwise division
v1 * 2  # scalar multiplication

## universal function -> functional programming
def reverse_num(x):
    return 1/x

# broadcasting
for x in v1:
    print(reverse_num(x))
```

```python
## v1과 v2에 있는 모든 원소들끼리 뺀 다음에, 제곱하고 다 더해서 루트 씌운거. = Euclidean distance(L2 distance)

#(v1 - v2) ** 2
distance = np.sqrt(np.sum(np.square(v1 - v2)))
distance2 = np.linalg.norm(v1 - v2) # 벡터의 크기 (L2 norm)
distance
#distance == distance2
```

```python
f = lambda x,y : np.sqrt(np.sum(np.square(x, y)))
f(v1, v2)
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum(np.square(a - b)))
    #return np.linalg.norm(a - b)
```

```python
def cosine_similarity(a, b):
    # 1. 분자 : a o b
    dot = a @ b

    # 2. 분모 : ||a||, ||b||
    norm = np.linalg.norm(a) * np.linalg.norm(b) + 1e-12
    sim = dot / norm

    if np.isnan(sim):
        return 0
    else:
        return sim
```

```python
print(euclidean_distance(v1, v2))
print(cosine_similarity(v1, v2))
```

```python
## euclidean distance version

target = data.loc[2] # 6번 고객
distances = []

## target을 기준으로 모든 데이터에 대해서 유사도를 계산한 뒤, 가장 유사한 고객을 찾아주세요. (index 출력.)
# euclidean -> cosine
for idx, row in data.iterrows():
    distances.append(euclidean_distance(target, row))
```

```python
distances[2] = 999999 # 6번 고객만 적당한 큰 값으로 변경.
```

```python
np.argmin(distances) # 18154
```

```python
## cosine similarity version

target = data.loc[2] # 6번 고객
similarities = []

## target을 기준으로 모든 데이터에 대해서 유사도를 계산한 뒤, 가장 유사한 고객을 찾아주세요. (index 출력.)
#  cosine
for idx, row in data.iterrows():
    similarities.append(cosine_similarity(target, row))
```
